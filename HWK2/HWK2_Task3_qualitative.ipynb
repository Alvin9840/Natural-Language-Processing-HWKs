```python
import pandas as pd

# Load predictions from all four notebooks
zero_shot_df = pd.read_csv("zero_shot_predictions.csv")
handpicked_df = pd.read_csv("handpicked_predictions.csv")
random_df = pd.read_csv("random_predictions.csv")
similarity_df = pd.read_csv("similarity_predictions.csv")

# print(f"Zero-shot shape: {zero_shot_df.shape}")
print(f"Handpicked shape: {handpicked_df.shape}")
print(f"Random shape: {random_df.shape}")
print(f"Similarity shape: {similarity_df.shape}")
# print(f"\nZero-shot columns: {zero_shot_df.columns.tolist()}")
print(f"Handpicked columns: {handpicked_df.columns.tolist()}")
```

```python
# Merge all predictions into a single DataFrame
merged = zero_shot_df.copy()

# Add handpicked predictions (K=1 to K=5)
for K in range(1, 6):
    merged[f"pred_hand_k{K}"] = handpicked_df[f"pred_hand_k{K}"]

# Add random predictions (K=1 to K=5)
for K in range(1, 6):
    merged[f"pred_rand_k{K}"] = random_df[f"pred_rand_k{K}"]

# Add similarity predictions (K=1 to K=5)
for K in range(1, 6):
    merged[f"pred_sim_k{K}"] = similarity_df[f"pred_sim_k{K}"]

print(f"\nMerged shape: {merged.shape}")
print(f"\nMerged columns ({len(merged.columns)}):")
print(merged.columns.tolist())
print(f"\nFirst row:")
print(merged.iloc[0])
```

```python
# Helper function to display examples
def show_example(idx: int):
    row = merged[merged["idx"] == idx].iloc[0]
    print(f"Index: {int(row['idx'])}")
    print(f"Text:  {row['text']}\n")
    print(f"Gold label        : {row['role']}")
    print(f"Zero-shot         : {row['pred_zero_shot']}")
    print(f"Handpicked K=1    : {row['pred_hand_k1']}")
    print(f"Random K=3        : {row['pred_rand_k3']}")
    print(f"Similarity K=5    : {row['pred_sim_k5']}")
    print("-" * 80)
```

```python
# Victim
victim_candidates = merged[
    (merged["role"] == "Victim") &
    (merged["pred_zero_shot"] != "Victim") &
    (merged["pred_sim_k5"] == "Victim")
]

print(f"Found {len(victim_candidates)} Victim examples (zero-shot wrong, similarity correct)\n")
print(victim_candidates[["idx", "pred_zero_shot", "pred_sim_k5"]].head(10))

print("\n" + "="*80)
print("SHOWING FIRST 3 VICTIM EXAMPLES")
print("="*80 + "\n")

for idx in victim_candidates["idx"].head(3):
    show_example(idx)
```

```python

# Third
third_candidates = merged[
    (merged["role"] == "Third") &
    (merged["pred_zero_shot"] != "Third") &
    (merged["pred_hand_k1"] == "Third")
]

print(f"Found {len(third_candidates)} Third examples (zero-shot wrong, handpicked K=1 correct)\n")
print(third_candidates[["idx", "pred_zero_shot", "pred_hand_k1"]].head(10))

print("\n" + "="*80)
print("SHOWING FIRST 3 THIRD EXAMPLES")
print("="*80 + "\n")

for idx in third_candidates["idx"].head(3):
    show_example(idx)
```

```python
# Abuser
abuser_candidates = merged[
    (merged["role"] == "Abuser") &
    (merged["pred_zero_shot"] == "Abuser") &
    (merged["pred_sim_k1"] == "Abuser") &
    (merged["pred_hand_k1"] == "Abuser")
]

print(f"Found {len(abuser_candidates)} Abuser examples (all methods correct)\n")
print(abuser_candidates[["idx", "pred_zero_shot", "pred_hand_k1", "pred_sim_k1"]].head(10))

print("\n" + "="*80)
print("SHOWING FIRST 3 ABUSER EXAMPLES")
print("="*80 + "\n")

for idx in abuser_candidates["idx"].head(3):
    show_example(idx)
```
